```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data_df = pd.merge(city_df, ride_df, how='inner', on='city')

# Display the data table for preview
combined_data_df.head()
#city_type = combined_data_df.groupby(['type','city'])

# Find out how many of each gender took bike trips
#sum_it_up = city_type.sum()

#sum_it_up
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_groups = combined_data_df.groupby(['type'])
print(type_groups)
fare_sum = type_groups["fare"].sum()
################################################################
           
city_fare_table = pd.DataFrame({"Fare": fare_sum})
#city_fare_table

#city_fare_table["Fare Percentage"] = city_fare_table["Fare"] / fare_total
#city_fare_table
#city_fare_table["Fare Percentage"] = city_fare_table["Fare Percentage"].map("{:,.2}".format)
#city_fare_table
#city_summary_table
##############################################################
# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = city_fare_table["Fare"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=45)
plt.axis("equal")
plt.title("% of Total Fares by City Type", fontsize=16)


# Save Figure
plt.savefig("Images/TotalFaresPie.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rural_df = combined_data_df.loc[combined_data_df["type"] == "Rural", :]
#rural_fare_total = rural_df["fare"].sum()
suburban_df = combined_data_df.loc[combined_data_df["type"] == "Suburban", :]
#suburban_fare_total = suburban_df["fare"].sum()
urban_df = combined_data_df.loc[combined_data_df["type"] == "Urban", :]
#urban_fare_total = urban_df["fare"].sum()
################
rural_rides = sum(rural_df["city"].value_counts())
suburban_rides = sum(suburban_df["city"].value_counts())
urban_rides = sum(urban_df["city"].value_counts())

# Build Pie Chart
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [rural_rides, suburban_rides, urban_rides]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=0)
plt.axis("equal")
plt.title("% of Total Rides by City Type", fontsize=16)


# Save Figure
plt.savefig("Images/TotalRidesPie.png")
```
